```python
set1={1,2,3,48,96}
set2={4,5,6,25,48}
set3=set1.union(set2)
set4=set1.intersection(set2)
print(set3)
print(set4)
```

```python
set1={1,2,3,48,96}
set2={4,5,6,25,48}
# print(set1,set2)     
set2.intersection_update(set1)
print(set1,set2)
```

```python
# difference--gives element of first position and removes common from them
set1={1,2,3,48,96}
set2={4,5,6,25,48}
set3=set2.difference(set1)
print(set3)
```

```python
set1={1,2,3,48,96}
set2={4,5,6,25,48}
print(set1,set2)
set1.difference_update(set2)
print(set1,set2)
```

```python
# isdisjoint()--return true when no common lement and returns false when
# common element is found
set1={1,2,3,4,5}
set2={9,8,7,6}
set1.isdisjoint(set2)
```

```python
# symmetric_difference()--union of sets without common elements
set1={1,2,3,4,5}
set2={4,5,6,7,8}
set3=set1.symmetric_difference(set2)
# set1.symmetric_difference_update(set2)
print(set1)
```
