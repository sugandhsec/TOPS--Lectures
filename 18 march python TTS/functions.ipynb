```python
# positional 
def fuc1(a,b):
  print(a,b)
fuc1("hello",45)  
```

```python
#keyword argument
def fuc1(a,b,c):
  print(a,b,c)
fuc1(89,c="ram",b="sitaji")  
```

```python
#defualt argument
def fuc1(a,b=45):
  print(a,b)
fuc1("hello")  
```

```python
#arbitry argument
def fuc1(a,b,*c):
  print(a,b,c)
fuc1(45,56,78,88,223,445,)  
```

```python
#arbitry keyword argument
def fuc1(a,b,**c):
  print(a,b,c)
fuc1(12,b=12,c="ram",e=5,d=5,f=6)  
```
