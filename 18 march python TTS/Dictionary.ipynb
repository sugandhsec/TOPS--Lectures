# dictionary methods

```python
dict1={1:"hello",2:"world",True:"wow",3:"Last"}
print(dict1)
```

```python
# keys()--gives list of keys in dictionary
dict1={1:"World",2:"World"}
x=dict1.keys()
print(x)
```

```python
# Values()--gives list of values in dictionary
dict1={1:"World",2:"Now"}
x=dict1.values()
print(x)
```

```python
# items()--gives list of tuple  and tuple have two value one key and another value
dict1={1:"World",2:"Now"}
x=dict1.items()
print(x)
```

```python
dict1={1:"World",2:"Now"}
# dict1.pop(45)---KeyError
dict1.pop(1)
# dict1.pop() ---in this pop() can't be blank
print(dict1)
```

```python
dict1={1:"World",2:"Now"}
dict1.popitem()
print(dict1)
```

```python
dict1={1:"World",2:"Now"}
# print(dict1.get(45))
print(dict1.get(15,"Wrong index"))
print(dict1)
```

```python
dict1={1:"World",2:"Now"}
print(dict1.setdefault(45,"hello"))
print(dict1)
```

```python
dict1={1:"World",2:"Now"}
dict1.update({48:"True",56:"now"})
print(dict1)
```
