```python
# generators and iterators
# generator function
def fun1():
    list1=[1,2,3,4,5]
    for i in list1:
        yield i**2
a=fun1() #a is generator object ---- iterator
iter(a)
print(next(a))
print(next(a))
print(next(a))
for i in a:
    print(i)
```

```python
class one:
    def __init__(self):
        self.name=""
        self._age=0
#         print("Constrcutor Called")
    
    def takevalue(self):
        self.name=input("Enter Your Name: ")
        self.age=int(input("Enter Your Age: "))
    
    def printvalue(self):
        print(f"Your Name is {self.name}\n Your Age is {self.age}")
        
class two(one):
    def __init__(self):
        self.email=""
        self.mobile=""
        super().__init__()
    def getvalue(self):
        self.takevalue()
        self.email=input("Enter Your Email: ")
        self.mobile=input("Enter Your Mobile: ")
        
    def showvalue(self):
        self.printvalue()
        print(f"Your Email is {self.email}\n Your Mobile is {self.mobile} {self.age}")
#         _clasname__variablename
        
```
